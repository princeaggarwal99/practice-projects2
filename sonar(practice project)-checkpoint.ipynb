```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.22,random_state=45)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
xpred=lg.predict(x_test)
accuracy_score(xpred,y_test)
```

```python
confusion_matrix(xpred,y_test)
```

```python
classification_report(xpred,y_test)
```

```python
gb=GaussianNB()
```

```python
gb.fit(x_train,y_train)
x_pred=gb.predict(x_test)
accuracy_score(x_pred,y_test)
```

```python
svc=SVC()
svc.fit(x_train,y_train)
xsvcpred=svc.predict(x_test)
accuracy_score(xsvcpred,y_test)
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
xpre=dt.predict(x_test)
accuracy_score(xpre,y_test)
```
